# Regression

```python
dfr = df[['total_bill', 'tip']].copy()
dfr.rename(columns = {'total_bill':'x (total_bill)', 'tip':'y (tip)'}, inplace=True)
dfr['x-mean'] = np.mean(dfr['x (total_bill)'])
dfr['x-xbar'] = dfr['x (total_bill)'] - np.mean(dfr['x (total_bill)'])
dfr['y-ybar'] = dfr['y (tip)'] - np.mean(dfr['y (tip)'])
dfr['x-xbar^2'] = np.square(dfr['x-xbar'])
dfr['y-ybar^2'] = np.square(dfr['y-ybar'])
dfr['(x-xbar)*(y-ybar)'] = dfr['x-xbar']*dfr['y-ybar']
dfr.head()
```

```python
ss_x = np.sum(dfr['x-xbar^2'])
ss_y = np.sum(dfr['y-ybar^2'])
s_xy = np.sum(dfr['(x-xbar)*(y-ybar)'])
r = np.sum(s_xy)/np.sqrt(ss_x*ss_y)
sd_x = np.sqrt(ss_x/(dfr.shape[0]-1))
sd_y = np.sqrt(ss_y/(dfr.shape[0]-1))
slope = r*sd_y/sd_x
intercept = np.mean(dfr['y (tip)'])-slope*np.mean(dfr['x (total_bill)'])
```

```python
dfr['Predicted Tip'] = intercept + slope * dfr['x (total_bill)']


dfr['Residual'] = dfr['Predicted Tip'] - dfr['y (tip)']
dfr['RegSq'] = np.square(dfr['Predicted Tip'] - np.mean(dfr['Predicted Tip']))
dfr['ResSq'] = np.square(dfr['Residual'] - np.mean(dfr['Residual']))
dfr.head(10)
```

```python
RegSS = np.sum(dfr['RegSq'])
ResSS = np.sum(dfr['ResSq'])
TotalSS = RegSS+ResSS
Rsq = RegSS/TotalSS
RegDof = 1 # Total variables -1
ResDof = df.shape[0]-RegDof-1
TotalDof = RegDof+ResDof
AdjRsq = 1-((1-Rsq)*(TotalDof/ResDof))
StdErr = np.sqrt(ResSS/ResDof)
```

```python
ParamValue = [RegSS, ResSS, TotalSS, Rsq, RegDof, ResDof, TotalDof, AdjRsq, StdErr,slope,intercept]
ParamName = ['Regression Sum of Square',
             'Residual Sum of Square',
             'Total Sum of Squares',
             'R Square',
             'Regression Degrees of Freedom',
             'Residual Degrees of Freedom',
             'Total Degrees of Freedom',
             'Adjusted R Square',
             'Standard Error',
             'slope',
             'intercept']

stats = pd.DataFrame(data=ParamValue,index=ParamName,columns=['statistic value'])
print(stats)
```
